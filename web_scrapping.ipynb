```python
import requests
from bs4 import BeautifulSoup
headers = {'User-Agent' : "Mozilla/5.0 (Windows NT 6.1; Win64; x64)"}
url="https://www.justdial.com/Chennai/24-Hours-Taxi-Services/nct-10000404/page-1"
res = requests.get(url,headers=headers)
soup = BeautifulSoup(res.content,'html.parser')
print(soup.prettify())
```

```python
print(soup.find_all('li'))
```

```python
rows=soup.find_all('h2',class_='jcn')
```

```python
print(rows)
```

```python
print(type(rows[0]))
```

```python
print(rows[0].findAll('h2')[0].getText())
```

```python
for row in rows:
    print(row.findAll('li')[0].getText())
```
